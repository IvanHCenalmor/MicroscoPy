```python
import os 
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
import tensorflow as tf


def norm(img):
     return (img - img.min()) / (img.max() - img.min() + 1e-10)
 
def print_info(data, text):
     print(f'Information for {text}:')
     print(f'\tMin: {data.min()}')
     print(f'\tMax: {data.max()}')
     print(f'\tMean: {data.mean()}')
     print(f'\tType: {data.dtype}')
     print(f'\tShape: {data.shape}')

     plt.hist(data.flatten())
     plt.title('Histogram')
     plt.show()

img = io.imread('../datasets/EM/test/test-input0000.png')
print_info(img, 'original_image')

norm_img =  norm(img)
print_info(norm_img, 'normalized_image')

io.imsave('tf_img.tif', np.expand_dims(norm_img, -1))
read_tf_img = io.imread('tf_img.tif')
print_info(read_tf_img, 'post_saved_image')
```

```python
norm_img =  norm(img)
norm_img -= 0.2
print_info(norm_img, 'normalized_image minus 0.2')

io.imsave('tf_img.tif', np.expand_dims(norm_img, -1))
read_tf_img = io.imread('tf_img.tif')
print_info(read_tf_img, 'post_saved_image minus 0.2')
```

```python
norm_img =  norm(img)
norm_img = np.clip(norm_img, 0.2, 0.8)
print_info(norm_img, 'normalized_image clipped 0.2-0.8')

io.imsave('tf_img.tif', np.expand_dims(norm_img, -1))
read_tf_img = io.imread('tf_img.tif')
print_info(read_tf_img, 'post_saved_image clipped 0.2-0.8')
```
